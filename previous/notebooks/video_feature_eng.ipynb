```python
import pandas as pd 

import numpy as np
```

```python
df = pd.read_csv(r"E:\resume_project\ai-interview-analyzer\data\processed\training_features.csv")
```

```python
df = df.drop("image_name",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("emotion",axis=1)
y = df["emotion"]
```

```python
from sklearn.preprocessing import LabelEncoder ,StandardScaler

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print(np.unique(y_encoded))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

Model = RandomForestClassifier(
    n_estimators=700,
    max_depth=30,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)

Model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred_train = Model.predict(X_train_scaled)
y_pred_test = Model.predict(X_test_scaled)

print("Train Accuracy:", accuracy_score(y_train, y_pred_train))
print("Test Accuracy:", accuracy_score(y_test, y_pred_test))

print(classification_report(y_test, y_pred_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

Model = GradientBoostingClassifier(
    learning_rate=0.1,
    max_depth=5,
    n_estimators=300,
    random_state=42
)

Model.fit(X_train_scaled, y_train)

y_pred_train = Model.predict(X_train_scaled)
y_pred_test = Model.predict(X_test_scaled)

print("Train Accuracy:", accuracy_score(y_train, y_pred_train))
print("Test Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils.class_weight import compute_class_weight
from xgboost import XGBClassifier
from scipy.stats import randint, uniform

# Build class weights from training labels, then upweight classes with lowest recall.
classes = np.unique(y_train)
base_weights = compute_class_weight(class_weight="balanced", classes=classes, y=y_train)
class_weight_map = {int(c): float(w) for c, w in zip(classes, base_weights)}

# Extra emphasis for minority / hard classes: angry(0), disgust(1), fear(2)
for cls in [0, 1, 2]:
    if cls in class_weight_map:
        class_weight_map[cls] *= 1.25

sample_weight_train = np.array([class_weight_map[int(label)] for label in y_train], dtype=np.float32)

rf = RandomForestClassifier(random_state=42, n_jobs=-1)
rf_param_dist = {
    "n_estimators": randint(300, 1000),
    "max_depth": [None, 10, 20, 30, 40],
    "min_samples_split": randint(2, 12),
    "min_samples_leaf": randint(1, 6),
    "max_features": ["sqrt", "log2", None],
    "bootstrap": [True, False],
    "class_weight": [class_weight_map, "balanced", "balanced_subsample"],
}

xgb = XGBClassifier(
    objective="multi:softmax",
    num_class=len(np.unique(y_train)),
    eval_metric="mlogloss",
    random_state=42,
    n_jobs=-1,
)
xgb_param_dist = {
    "n_estimators": randint(200, 900),
    "max_depth": randint(3, 10),
    "learning_rate": uniform(0.02, 0.23),
    "subsample": uniform(0.65, 0.35),
    "colsample_bytree": uniform(0.65, 0.35),
    "min_child_weight": randint(1, 8),
    "gamma": uniform(0.0, 0.4),
    "reg_alpha": uniform(0.0, 0.4),
    "reg_lambda": uniform(0.7, 1.8),
}

rf_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=rf_param_dist,
    n_iter=12,
    scoring="f1_macro",
    cv=3,
    verbose=0,
    random_state=42,
    n_jobs=-1,
)

xgb_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=xgb_param_dist,
    n_iter=12,
    scoring="f1_macro",
    cv=3,
    verbose=0,
    random_state=42,
    n_jobs=-1,
)

rf_search.fit(X_train_scaled, y_train)
xgb_search.fit(X_train_scaled, y_train, sample_weight=sample_weight_train)

print("Class weight map:", class_weight_map)
print("Best RF CV Macro-F1:", rf_search.best_score_)
print("Best RF Params:", rf_search.best_params_)
print("Best XGB CV Macro-F1:", xgb_search.best_score_)
print("Best XGB Params:", xgb_search.best_params_)
```

```python
best_rf = rf_search.best_estimator_
best_xgb = xgb_search.best_estimator_

rf_test_pred = best_rf.predict(X_test_scaled)
xgb_test_pred = best_xgb.predict(X_test_scaled)

print("===== Tuned RandomForest =====")
print("Test Accuracy:", accuracy_score(y_test, rf_test_pred))
print(classification_report(y_test, rf_test_pred, target_names=label_encoder.classes_))

print("===== Tuned XGBoost =====")
print("Test Accuracy:", accuracy_score(y_test, xgb_test_pred))
print(classification_report(y_test, xgb_test_pred, target_names=label_encoder.classes_))
```
